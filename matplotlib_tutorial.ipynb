## **Introduction To Matplotlib**
### What is Matplotlib?
* Matplotlib is a popular Python library for creating static, animated, and interactive visualizations.
* It provides a comprehensive set of tools for creating high-quality 2D and 3D plots, charts, and graphs, making it a widely-used data visualization library in data science and scientific computing.
### Why Matplotlib?
1. **Visualization Library:** Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python.

2. **Plotting Capabilities:** Matplotlib provides a wide range of plotting capabilities, including line plots, scatter plots, histograms, bar charts, and more.

3. **Customization Options:** Matplotlib offers extensive customization options, allowing users to control colors, fonts, labels, titles, and other plot elements.

4. **Integration with Other Libraries:** Matplotlib integrates well with other popular Python libraries, including NumPy, Pandas, Scikit-learn, and Seaborn.

5. **Output Formats:** Matplotlib supports various output formats, including PNG, PDF, EPS, SVG, and more, making it easy to share and publish visualizations.

```python
# importing the libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### **Line Graph**
* A line graph, also known as a line plot or line chart, is a type of graph that displays data as a series of connected points, creating a line. 
* In Matplotlib, you can create a line graph using the plot() function.
* **plt.plot()** is a function in Matplotlib that creates a line plot of one or more sets of data.
* **Syntax:**

   **plt.plot(x, y, 'format_string', kwargs)**
*  Where x and y are the coordinates of the data points to be plotted.
 'format_string' is a string that specifies the line style, marker type, and color.
 kwargs are additional keyword arguments that can be used to customize the plot.



```python
x=[1,2,3,4,5]
y=[2,4,6,8,10]
plt.plot(x,y,label='2x',color='r',linewidth=2,marker='.',markersize=10,markeredgecolor='blue')
plt.title("Our First Graph !") # to add title to graph
plt.xlabel("X-axis") # to add x label to graph
plt.ylabel("Y-axis") # to add y label to graph
x2=np.arange(0,4.5,0.5)
plt.plot(x2,x2**2,label='2^x')
plt.legend() #displays legend
plt.savefig('mygraph.png',dpi=100) # saves picture of graph in current directory
plt.show()
```

### **Bar Chart**
* A bar chart, also known as a bar graph or bar plot, is a type of graph that displays data as rectangular bars. 
* In Matplotlib, you can create a bar chart using the bar() function.


```python
labels=['A','B','C']
values=[1,4,2]
bars=plt.bar(labels,values,color='#3CCDF6')
bars[0].set_hatch('/')
plt.figure(figsize=(6,4))
plt.show()
```

## **Gas Prices Plot**

```python
gas=pd.read_csv('gas_prices.csv')
gas.head()
```

```python
plt.title("Gas Prices Over Time",fontdict={'fontweight':'bold','fontsize':18,'color':'g'})
plt.xlabel('Year',fontdict={'fontweight':'bold','fontsize':12})
plt.ylabel('Price in USD',fontdict={'fontweight':'bold','fontsize':12})
plt.plot(gas['Year'],gas['USA'],label='United States',marker='.',markeredgecolor='b',color='r')
plt.plot(gas['Year'],gas['Australia'],label='Australia',marker='.',markeredgecolor='g')
plt.plot(gas['Year'],gas['UK'],label='United Kingdom',marker='.',markeredgecolor='b')
plt.legend()
plt.xticks(gas.Year[::3])
plt.savefig('gas_prices.png')
plt.show()


```

#### Loading Fifa Data

```python
fifa = pd.read_csv('fifa_data.csv')
fifa.head(3)
```

### **Histograms**

```python
bins=[40,50,60,70,80,90]
plt.title("Distribution Of Player Skills in FIFA 2023")
plt.hist(fifa.Overall,bins=bins)
plt.xlabel('Skill Level')
plt.ylabel('Number Of Players')
plt.show()
```

### **Pie Chart**

```python
vehicles=[15,25,30,80]
explodes=(.1,.1,0,0)
labels=['buses','cars','auto-rikshaws','bikes']
plt.pie(vehicles,labels=labels,autopct='%.2f',explode=explodes)
plt.title("Vehicles",fontdict={'fontweight':'bold','color':'g','fontsize':16})

# plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0)
plt.show()
```

### **Box Plots**

```python
cn=[70,72,55,56,45,51,48,61,63,68]
flat=[70,61,59,69,51,63,74,78,77,69]
ai=[81,78,79,75,68,61,47,59,58,60]
plt.title("Comparision Of Two Subjects",fontdict={'fontweight':'bold','color':'r'})
labels=['Computer Networks','FLAT','AI']
plt.ylabel("Marks Out Of 100")
plt.boxplot([cn,flat,ai],tick_labels=labels)
plt.show()
```
